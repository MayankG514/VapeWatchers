```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# 1. Distribution of Vape Types
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Vape_Type', order=df['Vape_Type'].value_counts().index)
plt.title("Distribution of Vape Types")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Youth Appeal Score by Brightness Level
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Brightness_Level', y='youth_appeal_score')
plt.title("Youth Appeal Score by Brightness Level")
plt.tight_layout()
plt.show()

# 3. Readability Youth Score by Vape Type
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Vape_Type', y='readability_youth_score', estimator='mean')
plt.title("Average Readability Youth Score by Vape Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Special Characters Youth Score by Contains Warning
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='contains_warning', y='special_chars_youth_score')
plt.title("Special Chars Youth Score by Warning Presence")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Vape_Type', hue='contains_emojis')
plt.title("Emoji Presence by Vape Type")
plt.xlabel("Vape Type")
plt.ylabel("Count")
plt.legend(title="Contains Emojis")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='contains_warning')
plt.title("Distribution of Warnings in Images")
plt.xlabel("Contains Warning")
plt.ylabel("Number of Images")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
cols = ['youth_appeal_score', 'readability_youth_score', 'special_chars_youth_score']
corr = df[cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Youth-Focused Scores")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='readability_youth_score', y='youth_appeal_score', hue='contains_warning')
plt.title("Youth Appeal vs. Readability Score")
plt.xlabel("Readability Youth Score")
plt.ylabel("Youth Appeal Score")
plt.legend(title="Warning Present")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='youth_appeal_score', bins=30, kde=True, color='purple')
plt.title("Distribution of Youth Appeal Scores")
plt.xlabel("Youth Appeal Score")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
